```python
def pokaz(ct_df):
    s = 0
    n = 0
    mx = 0
    mp = 0
    sstr_pr = [ ]
    for item,jtem in zip(ct_df['clean_price'],ct_df['clean_qv']):
        try:
            n += 1
            if float(jtem) > mx:
                mx = float(jtem)
                mp = float(item)
            sstr_pr.append(float(item)/float(jtem))
        except:
            pass
    sstr_pr = np.array(sstr_pr)   
    sstr_pr = sstr_pr[(sstr_pr < float(pd.DataFrame(sstr_pr).quantile(0.9))) & (sstr_pr > float(pd.DataFrame(sstr_pr).quantile(0.01)))]
    md = np.median(sstr_pr)
    s = np.mean(sstr_pr)
    n = len(sstr_pr)
    #if len(sstr_pr) == 0:
    #    return 0,0,0,0
    return mx,md,s,n
    #return sstr_pr

def counter(city):
    ct_df = preitog_table[preitog_table.city == city]
    #q_up = ct_df['clean_price'].quantile(1)
    #q_down = ct_df['clean_price'].quantile(0)
    #ct_df = ct_df[(ct_df['clean_price']<q_up) & (ct_df['clean_price']>q_down)]
    #print(ct_df.shape)
    mx,med,s,n = pokaz(ct_df)
    
    mxa,meda,sa,na = pokaz(ct_df[ct_df['class'] == 'a'])
    mxb,medb,sb,nb = pokaz(ct_df[ct_df['class'] == 'b'])
    mxc,medc,sc,nc = pokaz(ct_df[ct_df['class'] == 'c'])
    mini_df = pd.DataFrame([city,n,mx,s,med,na,mxa,sa,meda,nb,mxb,sb,medb,nc,mxc,sc,medc]).T
    mini_df.columns = ['Город','Всего','Максима площ', 'Ср цена за год','Медианная цена за год',\
                                'Всего А','Максим площ А','Ср цена за год А','Медианная цена за год A',\
                                'Всего В','Максим площ В','Ср цена за год В','Медианная цена за год B',\
                                'Всего С','Максим площ C','Ср цена за год С','Медианная цена за год C']

    return(mini_df)  
```
